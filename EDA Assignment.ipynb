```python
Null_Data = (Application_Data.isnull().sum() / len(Application_Data)) * 100

# Filter columns with non-zero missing percentages
Null_Columns = Null_Data[Null_Data > 0]

# Display the columns with their corresponding missing percentages
Null_Columns
```

```python
Fill_Null_Columns = Null_Columns.index.tolist()
Fill_Null_Columns
```
